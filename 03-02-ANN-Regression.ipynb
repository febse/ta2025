# Artificial Neural Networks for Regression


```python
import numpy as np
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
n = 100
p = 1

np.random.seed(23)

X = np.random.randn(n, p)
e = np.random.randn(n)
beta = np.array([2])
y = 2 + X @ beta + e
y = (y - np.mean(y)) / np.std(y)
```

```python
linreg = LinearRegression()
linreg.fit(X, y)

print("Intercept:", linreg.intercept_)
print("Coefficient:", linreg.coef_)
```

```python
# Plot the predicted values

plt.scatter(X, y)
plt.plot(X, linreg.predict(X), color='red', label = r'$\hat{y}$')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
```

## Simple Linear Regression as a ANN

With one predictor variable the simple linear model predicts an outcome variable as a linear function of the predictor variable:

$$
\hat{y}_i = \hat{\beta}_0 + \beta_1 x_{i}
$$

Where the coefficients are chosen to minimize the mean squared error loss function.

$$
\text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
$$

A linear regression model is a special case of a neural network with no hidden layers, a linear activation function, and a mean squared error loss function. To be consistent with the literature on neural networks we will rename $\beta_1$ to $w$ (weights) and $\beta_0$ to $b$ (bias). The equation for the ANN model is
absolutely the same as for linear regression:

$$
\hat{y}_i = b + w x_{i}
$$

The usual visualization of this model is a single neuron with $p$ inputs and a single output. The neuron compute whe weighted sum of its inputs and passes this value to its output nodes. For linear regression the output of the is passed through an identity activation function $\mathbb{I}(z) = z$ (i.e. unchanged) to the output node which is the predicted value $\hat{y}_i$.

In the simplest case we have one input (feature) and one output (target variable):

$$
\hat{y}_i = \mathbb{I}(b + w x_i)
$$

So if we know the values of the weight and the bias, we can compute the predicted value for any input by simply multiplying the input by the weight and adding the bias. How do we find sensible values for the weight and the bias?



## Gradient Descent for Simple Linear Regression

{{< video ./videos/GradientDescentQuadratic.mp4 >}}


{{< video ./videos/GradientDescentLinearRegressionTwoParams.mp4 >}}


We can use gradient descent to find the optimal values of the weights and the bias that minimize the loss function. The loss function for linear regression is the mean squared error (MSE) which for a single observation (sample) is:

$$
L_i = (\text{actual} - \text{predicted})^2 = (y_i - \hat{y}_i)^2
$$

The gradient descent algorithm iteratively updates the weights and the bias in the direction of the negative gradient of the loss function with respect to the weights and the bias.

$$
\begin{align*}
w_1^{\text{new}} = w_1^{\text{old}} - \eta \frac{\partial L_i}{\partial w_1} \\
b^{\text{new}} = b^{\text{old}} - \eta \frac{\partial L_i}{\partial b}
\end{align*}
$$

where $\eta$ is the learning rate, a hyperparameter that controls the step size of the updates. To actually perform the updates, we need to compute the gradients of the loss function with respect to the weights and the bias. This is quite straightforward here:


$$
\begin{align*}
\frac{\partial L_i}{\partial w_1} & = \frac{\partial }{\partial w_1} (y_i - \hat{y}_i)^2 \\
                                  & = (y_i - \hat{y}_i) \frac{\partial }{\partial w_1} \hat{y}_i \\
                                  & = -2 (y_i - \hat{y}_i) x_{i}
\end{align*}
$$

$$
\begin{align*}
\frac{\partial L_i}{\partial b} & = \frac{\partial }{\partial b} (y_i - \hat{y}_i)^2 \\
                                  & = (y_i - \hat{y}_i) \frac{\partial }{\partial b} \hat{y}_i \\
                                  & = -2 (y_i - \hat{y}_i)
\end{align*}
$$

It is useful to also compute the update equations not just for a single observation, but for a subset (batch) of the observations.

The MSE loss for a batch of size m is:

$$
L = \frac{1}{m} \sum_{i=1}^{m} (y_i - \hat{y}_i)^2
$$

Exactly as before, we differentiate the loss with respect to the weight and the bias to get the update equations for the batch:

$$
\begin{align*}
\frac{\partial L}{\partial w} & = \frac{1}{m} \sum_{i=1}^{m} \frac{\partial }{\partial w} (y_i - \hat{y}_i)^2 \\
                                    & = \frac{1}{m} \sum_{i=1}^{m} (y_i - \hat{y}_i) \frac{\partial }{\partial w} \hat{y}_i \\
                                    & = -\frac{2}{m} \sum_{i=1}^{m} (y_i - \hat{y}_i) x_{i}                     
\end{align*}
$$               

It is also convenient to express these equations in vectorized form. It has the advantage that it maps nicely to matrix operations that we can express directly using `numpy`. Another advantage is that it generalizes easily to multiple input features and multiple output targets. Furthermore, vectorized operations are usually more computationally efficient. 

The sum involved in the gradient computation above is just the dot product of the column vectors of the errors $(y - \hat{y})$ and the inputs $x$:

$$
\begin{align*}
\frac{\partial L}{\partial w} & = -\frac{2}{m} (y - \hat{y})^T x \\
\frac{\partial L}{\partial b} & = -\frac{2}{m} \sum_{i=1}^{m} (y_i - \hat{y}_i)
\end{align*}
$$

or equivalently:

$$
\begin{align*}
\frac{\partial L}{\partial w} & = -\frac{2}{m} x^T (y - \hat{y}) \\
\frac{\partial L}{\partial b} & = -\frac{2}{m} \sum_{i=1}^{m} (y_i - \hat{y}_i)
\end{align*}
$$


```python
# Hyperparameters
epochs = 200
lr = 0.05

# First we initialize the parameters to 
# random values close to zero

w = np.random.randn(p).reshape(-1, p)
b = np.random.randn(1)

losses = []

# Training loop
for epoch in range(epochs):
    # Forward pass
    y_pred = X @ w + b
    
    # Compute loss (MSE)
    loss = np.mean((y.reshape(-1, 1) - y_pred) ** 2)
    losses.append(loss)
    
    # Backward pass (compute gradients)
    errors = y.reshape(-1, 1) - y_pred
    gradient_w = -2 * X.T @ errors / n
    gradient_b = -2 * errors.mean()
    
    # Update parameters
    w = w - lr * gradient_w
    b = b - lr * gradient_b

# Print final parameters
print("Weight:", w.item())
print("Bias:", b.item())

# Plot training loss
plt.plot(losses)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training Loss over Epochs (NumPy)')
plt.show()

# Plot predictions
y_pred_final = X @ w + b

plt.scatter(X, y, label='Data')
plt.plot(X, y_pred_final, color='red', label='Prediction')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

## A Neural Network and Function Approximation

Neural networks become more interesting when we add hidden layers and non-linear activation function.

### Activation Functions

The hidden layers of the neural network enable it to capture non-linear relationships between the input features and the target variable. This is achieved by applying non-linear activation functions to the outputs of the neurons in the hidden layers. Common activation functions include the ReLU (Rectified Linear Unit), sigmoid, and tanh functions.

$$
ReLU(z) = \max(0, z)
$$

$$
\sigma(z) = \frac{1}{1 + e^{-z}}
$$

$$
\tanh(z) = \frac{e^{z} - e^{-z}}{e^{z} + e^{-z}}
$$


```python
# Plot ReLU, Sigmoid, and Tanh functions

x = np.linspace(-4, 4, 100)

relu = np.maximum(0, x)
sigmoid = 1 / (1 + np.exp(-x))
tanh = np.tanh(x)

plt.plot(x, relu, label='ReLU')
plt.plot(x, sigmoid, label='Sigmoid')
plt.plot(x, tanh, label='Tanh')
plt.axhline(-1, color='gray', linestyle='--', linewidth=0.5)
plt.axhline(0, color='gray', linestyle='--', linewidth=0.5)
plt.axhline(1, color='gray', linestyle='--', linewidth=0.5)
plt.ylim(-1.5, 1.5)
plt.xlabel('x')
plt.ylabel('Activation')
plt.title('Activation Functions')
plt.legend()
plt.show()
```

For a network with one one input and two neurons in the hidden layer, one output and ReLU activation function, the equation for the output is:

https://www.desmos.com/calculator/jwruoptwuv


$$
\hat{y}_i = w_{o1} \cdot \text{ReLU}(h_{i1}) + w_{o2} \cdot \text{ReLU}(h_{i2}) + b_o
$$

The hidden layer neurons compute their activations as:

$$
\begin{align*}
h_{i1} & = w_{1} \cdot x_{i} + b_{1} \\
h_{i2} & = w_{2} \cdot x_{i} + b_{2}
\end{align*}
$$

Again, the MSE is the average squared difference between the actual and predicted values:

$$
MSE = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
$$

So the final equation for the output is:

$$
\hat{y}_i = w_{o1} \cdot \underbrace{\text{ReLU}(\overbrace{w_{1} \cdot x_i + b_{1}}^{h_{i1}})}_{z_{i1}} + w_{o2} \cdot \underbrace{\text{ReLU}(\overbrace{w_{2} \cdot x_i + b_{2}}^{h_{i2}})}_{z_{i2}} + b_o
$$

https://www.desmos.com/calculator/gmsdth4fmf

https://www.desmos.com/calculator/xnywykcn3l


It is convenient to rewrite the whole model with more compact (matrix) notation because
this will directly map to `numpy` code later on. It will also generalize easily to multiple input features and multiple output targets.

The network flow starts with the input vector $x$ with n samples (observations) and $p$ features (for now $p = 1$). Let's write some random values for the input vector and limit the number of samples to $n = 2$ for simplicity:

$$
x = \begin{bmatrix}
x_{i=1} \\
x_{i=2}
\end{bmatrix} = \begin{bmatrix}
4 \\
2
\end{bmatrix}
$$

The input input vector is mapped to the hidden layer by a linear (strictly speaking affine) transformation (i.e. a matrix multiplication plus a bias addition):

$$
H = x W_h + b_h
$$

The dimension of the weight matrix is determined by the number of input features and the number of neurons in the hidden layer. In our case we have $p = 1$ input feature and $h = 2$ neurons in the hidden layer, so the weight matrix has dimension $1 \times 2$:

$$
W_h = \begin{bmatrix}
w_{1} & w_{2}
\end{bmatrix} = \begin{bmatrix}
0.25 & 0.5
\end{bmatrix}
$$

The bias vector has dimension $1 \times h$:

$$
b_h = \begin{bmatrix}
b_{1} & b_{2}
\end{bmatrix} = \begin{bmatrix}
0.1 & -1.1
\end{bmatrix}
$$

```python
x = np.array([[4], [2]])
print("Input matrix shape:", x.shape)
print(x)

W_h = np.array([[0.25, 0.5]]).reshape(-1, 2)
print("Hidden layer weights shape:", W_h.shape)
print(W_h)

b_h = np.array([0.1, -1.1]).reshape(-1, 2)
print("Hidden layer biases shape:", b_h.shape)
print(b_h)
```


So the transformation from the input layer to the hidden layer is:

$$
H = \begin{bmatrix}
4 &
2
\end{bmatrix} \cdot \begin{bmatrix}
0.25 \\ 0.5
\end{bmatrix} + \begin{bmatrix}
0.1 & -1.1 \\
0.1 & -1.1
\end{bmatrix} = 
\begin{bmatrix}
4 \cdot 0.25 + 0.1 & 4 \cdot 0.5 - 1.1 \\
2 \cdot 0.25 + 0.1 & 2 \cdot 0.5 - 1.1
\end{bmatrix} =
\begin{bmatrix}
1.1 & 0.9 \\
0.6 & -0.1
\end{bmatrix}
$$

```python
# Compute hidden layer activations

H = x @ W_h + b_h
print("Hidden layer shape:", H.shape)
print(H)
```

Now we have the hidden layer activations and we need to pass them through the activation function (ReLU in this case). The ReLU function is applied element-wise to the hidden layer matrix H:

$$
Z = \text{ReLU}(H) = 
\begin{bmatrix}
\text{ReLU}(1.1) & \text{ReLU}(0.9) \\
\text{ReLU}(0.6) & \text{ReLU}(-0.1)
\end{bmatrix} =
\begin{bmatrix}
1.1 & 0.9 \\
0.6 & \mathbf{0}
\end{bmatrix}
$$

```python
def relu(x):
    return np.maximum(0, x)

Z = relu(H)
print("Activated hidden layer shape:", Z.shape)
print(Z)
```

The last step in our network is to map the hidden layer after activation ($Z$) to the output layer by another linear transformation:

$$
\hat{Y} = Z W_o + b_o
$$

Where the output weight matrix has dimension $h \times q$ where $q$ is the number of output targets (for now $q = 1$):

$$
W_o = \begin{bmatrix}
w_{o1} \\
w_{o2} \\
\end{bmatrix} = \begin{bmatrix}
0.4 \\
-0.6 \\
\end{bmatrix}
$$

The output bias vector has dimension $1 \times q$:
$$
b_o = \begin{bmatrix}
b_{o}
\end{bmatrix} = \begin{bmatrix}-0.2\end{bmatrix}
$$


```python
W_o = np.array([0.4, -0.6]).reshape(-1, 1)

print("Output layer weights shape:", W_o.shape)
print(W_o)

b_o = np.array([-0.2]).reshape(-1, 1)
print("Output layer biases shape:", b_o.shape)
print(b_o)
```

Now we are all set to actually compute the output predictions for our two samples (observations):

$$
\hat{Y} = 
Z W_o + b_o =
\begin{bmatrix}
1.1 & 0.9 \\
0.6 & 0
\end{bmatrix} \begin{bmatrix}
0.4 \\
-0.6
\end{bmatrix} + \begin{bmatrix}
-0.2 \\
-0.2
\end{bmatrix}
= \begin{bmatrix}
1.1 \cdot 0.4 + 0.9 \cdot -0.6 - 0.2 \\
0.6 \cdot 0.4 + 0 \cdot -0.6 - 0.2
\end{bmatrix} =
\begin{bmatrix}
-0.14 \\
0.04
\end{bmatrix}
$$

```python
y_pred = Z @ W_o + b_o
print("Output shape:", y_pred.shape)
print(y_pred)
```

If we have actual target values (for example the ones below) for our two samples, we can compute the MSE loss:

$$
y = \begin{bmatrix}
1.0 \\
0.0
\end{bmatrix}
$$

$$
MSE = \frac{1}{2} \left( (1.0 - -0.14)^2 + (0.0 - 0.04)^2 \right) = 0.5792
$$

## Weights Update Equations

Now that we have the predictions we also need to setup the update equations for the weights (the hidden layer weights and biases and the output layer weights and biases).

For this we need the gradients of the loss function with respect to all these parameters. Although this is actually quite easy, one can easily get lost in the derivation, especially when writing the vectorized versions.

### Output Layer Weights and Bias

Let's start with the gradients for the output layer weights and bias. The loss function is exactly the same as before:

$$
\text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
$$

$$
\begin{align*}
\frac{\partial L}{\partial w_{o1}} & = \frac{1}{m} \sum_{i=1}^{m} \frac{\partial }{\partial w_{o1}} (y_i - \hat{y}_i)^2 \\
                                  & = \frac{1}{m} \sum_{i=1}^{m} (y_i - \hat{y}_i) \frac{\partial }{\partial w_{o1}} \hat{y}_i \\
                                  & = -\frac{2}{m} \sum_{i=1}^{m} (y_i - \hat{y}_i) \frac{\partial }{\partial w_{o1}} \left( w_{o1} z_{i1} + w_{o2} z_{i2} + b_o \right) \\
                                  & = -\frac{2}{m} \sum_{i=1}^{m} (y_i - \hat{y}_i) z_{i1}
\end{align*}
$$

In the same way we can derive the gradient with respect to the second output weight:

$$
\begin{align*}
\frac{\partial L}{\partial w_{o2}} & = - \frac{2}{m} \sum_{i=1}^{m} (y_i - \hat{y}_i) z_{i2}
\end{align*}
$$



When we look at the two partial derivatives we can see that they are only a scaled (by $-\frac{2}{m}$) version of the dot product between the error vector $(y - \hat{y})$ and the respective hidden layer activation vector ($z_{1}$ or $z_{2}$). These vectors hold the data for all the samples.

Let's pack both partial derivatives into a single vectorized equation for the output layer weights:

$$
\underbrace{\frac{\partial L}{\partial W_o}}_{2 \times 1} = -\frac{2}{m} 
\begin{bmatrix}
\sum_{i=1}^{m} (y_i - \hat{y}_i) z_{i1} \\
\sum_{i=1}^{m} (y_i - \hat{y}_i) z_{i2}
\end{bmatrix} =
-\frac{2}{m} \underbrace{Z^T}_{2 \times m} \underbrace{(y - \hat{y})}_{m \times 1}
$$

:::{.callout-note}
## Shapes

Always pay close attention to the shapes of the matrices and vectors involved in the computations.

:::

### Hidden Layer Weights and Biases

As the model predictions depend on the hidden layer weights and biases through the hidden layer activations, we also need to update these during the training, again applying the chain rule for differentiation.

Before we start, let's define the derivative of the ReLU activation function:

$$
\text{ReLU}'(z) = \begin{cases}
1 & \text{if } z > 0 \\
0 & \text{otherwise}
\end{cases}
$$

You will note that $\text{ReLU}(x) = \max(0, x)$ does not have a derivative at $x = 0$. In practice we _assume_ that the derivative at this point is 0.

With this in mind we can go forward and differentiate the MSE loss with respect to the hidden layer weights and biases. Let's start with the first hidden layer weight $w_1$:

$$
\begin{align*}
\frac{\partial L}{\partial w_{1}} & = \frac{1}{m} \sum_{i=1}^{m} \frac{\partial }{\partial w_{1}} (y_i - \hat{y}_i)^2 \quad \text{just as before}\\
                                  & = \frac{1}{m} \sum_{i=1}^{m} (y_i - \hat{y}_i) \frac{\partial }{\partial w_{1}} \hat{y}_i \\
                                  & = -\frac{2}{m} \sum_{i=1}^{m} (y_i - \hat{y}_i) \frac{\partial }{\partial w_{1}} \left( w_{o1} z_{i1} + w_{o2} z_{i2} + b_o \right) \\
                                  & = -\frac{2}{m} \sum_{i=1}^{m} (y_i - \hat{y}_i) w_{o1} \frac{\partial }{\partial w_{1}} z_{i1} \quad \text{because the first weight enters the prediction only through } z_{i1} \\
                                  & = -\frac{2}{m} \sum_{i=1}^{m} (y_i - \hat{y}_i) w_{o1} \text{ReLU}'(h_{i1}) \frac{\partial }{\partial w_{1}} h_{i1} \quad \text{because z is h after activation} \\
                                  & = -\frac{2}{m} \sum_{i=1}^{m} (y_i - \hat{y}_i) w_{o1} \text{ReLU}'(h_{i1}) \frac{\partial }{\partial w_{1}} \left( w_{1} x_{i} + b_{1} \right) \quad \text{replace with the definition of h} \\
                                  & = -\frac{2}{m} \sum_{i=1}^{m} (y_i - \hat{y}_i) w_{o1} \text{ReLU}'(h_{i1}) x_{i} \\
                                  & = w_{o1} \left(-\frac{2}{m} \sum_{i=1}^{m} (y_i - \hat{y}_i) \text{ReLU}'(h_{i1}) x_{i}\right)
\end{align*}
$$

Running the same calculation for the second hidden layer weight $w_2$ we get:

$$
\begin{align*}
\frac{\partial L}{\partial w_{2}} & = w_{o2} \left(-\frac{2}{m} \sum_{i=1}^{m} (y_i - \hat{y}_i) \text{ReLU}'(h_{i2}) x_{i}\right)
\end{align*}
$$

In a similar fashion we can also derive the gradients for the hidden layer biases (not shown here). After some algebra we can summarize the gradients for the hidden layer weights in a single vectorized equation:

$$
\underbrace{\frac{\partial L}{\partial W_h}}_{p \times h} = -\frac{2}{m} X^T \left( \left( (y - \hat{y}) W_o^T \right) \odot \text{ReLU}'(H) \right)
$$

and for the hidden layer biases:

$$
\underbrace{\frac{\partial L}{\partial b_h}}_{1 \times h} = -\frac{2}{m} \sum_{i=1}^{m} \left( (y_i - \hat{y}_i) W_o^T \odot \text{ReLU}'(h_{i}) \right)
$$

## Implementing the Neural Network in NumPy


```python
# 1) We start by generating some data from a known function with added noise

np.random.seed(42)

# Set the number of data points (samples)
n = 100

# Generate input data, these will be evenly spaced points between -2 and 2
x = np.linspace(-2, 2, n)

def g_func(x):
    return x**3 + 2 * x ** 2 - 5 * x + 2

# Calculate the output data (with noise)
y = g_func(x) + np.random.normal(0, 0.5, size=n)

# Scale the data to zero mean and unit variance (important!)

x_scaled = (x - np.mean(x)) / np.std(x)
y_scaled = (y - np.mean(y)) / np.std(y)

# Sanity check: print means and stds
print("x mean:", np.mean(x), "std:", np.std(x))
print("y mean:", np.mean(y), "std:", np.std(y))

# Plot the data

plt.figure(figsize=(10, 5))
plt.scatter(x, y, alpha=0.5, s=20, label='Sampled data')
plt.plot(x, g_func(x), 'r-', label='True Function', linewidth=2)
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Neural network regression using NumPy with ReLU activation

# First we set the random seeds for reproducibility
# as these are used to initialize the weights

np.random.seed(42)
torch.manual_seed(42)

# How many hidden units
hidden_size = 3

# Initialize the weights and biases
Wh = np.random.randn(1, hidden_size) * np.sqrt(2.0 / 1)
bh = np.zeros((1, hidden_size))
Wo = np.random.randn(hidden_size, 1) * np.sqrt(2.0 / hidden_size)
bo = np.zeros((1, 1))

# Sanity check: print the shapes of the weights and biases
print("Wh shape:", Wh.shape)
print("bh shape:", bh.shape)
print("Wo shape:", Wo.shape)
print("bo shape:", bo.shape)

# Prepare the training data (need to be 2D arrays)

x_train_np = x_scaled.reshape(-1, 1)
y_train_np = y_scaled.reshape(-1, 1)

print("x_train_np shape:", x_train_np.shape)
print("y_train_np shape:", y_train_np.shape)

# Some helper function to make the code
# below easier to read

def relu(x):
    return np.maximum(0, x)

def relu_deriv(x):
    return (x > 0).astype(float)

# Training parameters
epochs = 1000
lr = 0.03
losses_np = []

# Training loop
for epoch in range(epochs):
    # Forward pass (calculate predictions)

    H = x_train_np @ Wh + bh
    Z = relu(H)
    y_pred = Z @ Wo + bo
    
    # Compute loss (MSE)
    loss = np.mean((y_train_np - y_pred) ** 2)
    losses_np.append(loss)
    
    # Backward pass
    n = len(y_train_np)
    dLoss_dy = - 2 * (y_train_np - y_pred) / n
    
    # Output layer gradients
    dWo = Z.T @ dLoss_dy
    dbo = np.sum(dLoss_dy)
    
    # Hidden layer gradients
    dZ = dLoss_dy @ Wo.T
    dH = dZ * relu_deriv(H) # Note the element-wise (not matrix) multiplication here
    dWh = x_train_np.T @ dH
    dbh = np.sum(dH, axis=0, keepdims=True)
    
    # Update weights and biases
    Wo -= lr * dWo
    bo -= lr * dbo
    Wh -= lr * dWh
    bh -= lr * dbh

# Final predictions (after the training loop)
H = x_train_np @ Wh + bh
Z = relu(H)

y_pred_np_final = Z @ Wo + bo

# Plot results
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.scatter(x_train_np, y_train_np, alpha=0.5, s=20, label='Noisy data')
plt.plot(x_train_np, y_pred_np_final, 'r-', label='NumPy NN (ReLU)', linewidth=2)
plt.xlabel('x')
plt.ylabel('y')
plt.title('NumPy Neural Network with ReLU')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(losses_np)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training Loss (NumPy)')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# Neural network regression using PyTorch

np.random.seed(42)
torch.manual_seed(42)

x_train = torch.from_numpy(x_scaled.reshape(-1, 1)).float()
y_train = torch.from_numpy(y_scaled.reshape(-1, 1)).float()

# Define the neural network
class TorchNN(nn.Module):
    def __init__(self, hidden_size=12):
        super(TorchNN, self).__init__()
        self.hidden_size = hidden_size

        self.input_to_hidden = nn.Linear(1, hidden_size)
        self.hidden_activation = nn.ReLU()
        self.output_layer = nn.Linear(hidden_size, 1)
    
    def forward(self, x):
        # Map the input to hidden layer
        x = self.input_to_hidden(x)
        # Then apply activation to the result
        x = self.hidden_activation(x)

        # Then map the activated hidden layer to output
        x = self.output_layer(x)

        return x

# Initialize model, loss, and optimizer
model = TorchNN(hidden_size=3)

criterion = nn.MSELoss()

optimizer = torch.optim.Adam(model.parameters(), lr=0.01)

# Training loop
epochs = 1000
losses = []

for epoch in range(epochs):
    # Forward pass
    y_pred = model(x_train)
    loss = criterion(y_pred, y_train)
    losses.append(loss.item())
    
    # Backward pass
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

# Get predictions
with torch.no_grad():
    y_pred_final = model(x_train)

# Plot results
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.scatter(x_train, y_train, alpha=0.5, s=20, label='Noisy data')
plt.plot(x_train, y_pred_final.numpy(), 'r-', label='NN prediction', linewidth=2)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Neural Network Approximation of Noisy Function')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(losses)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training Loss')
plt.grid(True)

plt.tight_layout()
plt.show()
```

## Multiple Regression

With two or more predictor variables (features) the multiple linear regression model predicts an outcome variable as a linear function of the predictor variables:

$$
\hat{y} = w_1 x_1 + w_2 x_2 + b
$$

It is convenient to write this equation in vectorized form:

Let $X$ be a $n \times p$ matrix of input features (currently $n = 1$ and $p = 2$):

$$
X = \begin{bmatrix}
x_{11} & x_{12} \\
x_{21} & x_{22} \\
\vdots & \vdots \\
x_{n1} & x_{n2}
\end{bmatrix}
\quad 

W = \begin{bmatrix}
w_{1} \\
w_{2} \\
\vdots \\
w_{p}
\end{bmatrix}
$$

and $W$ a $p \times 1$ vector of weights:

Then the equation for the multiple linear regression model is:

$$
\hat{y} = X W + b
$$

## Multiple Inputs and a Hidden Layer

We can extend the multiple inputs linear regression model by adding a hidden layer with non-linear activation functions as we did before for the single input case. In this case the prediction equation becomes:

$$
\underbrace{\hat{y}}_{1 \times 1} = \text{ReLU}(\underbrace{X}_{1 \times p} \underbrace{W_h}_{p \times h} + \underbrace{b_h}_{1 \times h}) \underbrace{W_o}_{h \times 1} + \underbrace{b_o}_{1 \times 1}
$$

## Tanh Activation

Replacing the ReLU activation function with $\tanh$ only changes the last equation:

$$
\begin{align*}
\hat{y} & = w_{o1} \tanh(h_{i1}) + w_{o2} \tanh(h_{i2}) + b_{o} \\
\end{align*}
$$

Let's derive the gradients with respect to the weights. The output layer is quite simple:

$$
\begin{align*}
\frac{\partial MSE}{\partial w_{o1}} & = \frac{1}{n} \sum_{i=1}^{n} \frac{\partial }{\partial w_{o1}} (y_i - \hat{y}_i)^2 \\
                                  & = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i) \frac{\partial }{\partial w_{o1}} \hat{y}_i \\
                                  & = -\frac{2}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i) \tanh(h_{i1})
\end{align*}
$$

So this partial derivative is a scaled dot product of the errors and the activations of the first hidden neuron. Similarly for the second output weight:

$$
\begin{align*}
\frac{\partial MSE}{\partial w_{o2}} & = - \frac{2}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i) \tanh(h_{i2})
\end{align*}
$$

We can pack these two equations into a whole gradient vector for the output weights:

$$
\frac{\partial MSE}{\partial w_{o}} = - \frac{2}{n} 
\begin{bmatrix}
\sum_{i=1}^{n} (y_i - \hat{y}_i) \tanh(h_{i1}) \\
\sum_{i=1}^{n} (y_i - \hat{y}_i) \tanh(h_{i2})
\end{bmatrix}
$$

or in vectorized form:

$$
\underbrace{\frac{\partial MSE}{\partial w_{o}}}_{2 \times 1} = - \frac{2}{n} 
\underbrace{Z^T}_{2 \times n} \underbrace{(y - \hat{y})}_{n \times 1}
$$

where Z is the matrix of hidden layer activations.

The gradients with respect to the hidden layer weights are a bit more involved because we need to apply the chain rule to account for the non-linear activation function. Let's derive the gradient with respect to the first hidden layer weight $w_1$. Before we do this, note that the derivative of the $\tanh$ function is:

$$
\frac{d}{d h} \tanh(h) = 1 - \tanh^2(h)
$$

$$
\begin{align*}
\frac{\partial MSE}{\partial w_{1}} & = \frac{\partial }{\partial w_{1}} \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2 \\
                                  & =   \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i) \frac{\partial }{\partial w_{1}} \hat{y}_i \\
                                  & = -\frac{2}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i) \frac{\partial }{\partial w_{1}} \left( w_{o1} \tanh(h_{i1}) + w_{o2} \tanh(h_{i2}) + b_{o} \right) \\
                                  & = -\frac{2}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i) w_{o1} \frac{\partial }{\partial w_{1}} \tanh(h_{i1}) \\
                                    & = -\frac{2}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i) w_{o1} \left( 1 - \tanh^2(h_{i1}) \right) \frac{\partial }{\partial w_{1}} h_{i1} \\
                                    & = -\frac{2}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i) w_{o1} \left( 1 - \tanh^2(h_{i1}) \right) x_{i} \\
\end{align*}
$$

For the second hidden layer weight $w_2$ the derivation is similar:

$$
\begin{align*}
\frac{\partial MSE}{\partial w_{2}} & = -\frac{2}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i) w_{o2} \left( 1 - \tanh^2(h_{i2}) \right) x_{i} \\
\end{align*}
$$
